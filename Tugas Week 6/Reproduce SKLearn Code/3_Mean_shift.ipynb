Tugas 6 Mean Shift Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift, estimate_bandwidth
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Membuat sampel data
centers = [[1, 1], [-1, -1], [1, -1]]
X, _ = make_blobs(n_samples=10000, centers=centers, cluster_std=0.6)

# Menggunakan Mean Shift untuk clustering
bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=500)
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
ms.fit(X)
ms_labels = ms.labels_
n_ms_clusters = len(np.unique(ms_labels))

# Menggunakan K-Means untuk clustering
kmeans = KMeans(n_clusters=3, n_init=10)
kmeans.fit(X)
kmeans_labels = kmeans.labels_
n_kmeans_clusters = len(np.unique(kmeans_labels))

# Plotting hasil clustering Mean Shift
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
markers = ["x", "o", "^"]
for k, marker in zip(range(n_ms_clusters), markers):
    cluster_center = ms.cluster_centers_[k]
    cluster_points = X[ms_labels == k]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], marker=marker)
    plt.scatter(cluster_center[0], cluster_center[1], marker=marker, color='black', s=200)
plt.title(f"Mean Shift - Estimated Clusters: {n_ms_clusters}")

# Plotting hasil clustering K-Means
plt.subplot(1, 2, 2)
markers = ["x", "o", "^"]
for k, marker in zip(range(n_kmeans_clusters), markers):
    cluster_center = kmeans.cluster_centers_[k]
    cluster_points = X[kmeans_labels == k]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], marker=marker)
    plt.scatter(cluster_center[0], cluster_center[1], marker=marker, color='black', s=200)
plt.title(f"K-Means - Estimated Clusters: {n_kmeans_clusters}")

plt.show()
```

Visualisasi output yang akan muncul akan terdiri dari dua subplot, masing-masing menampilkan hasil clustering dari dua algoritma, yaitu Mean Shift dan K-Means.

1. Mean Shift:
- Subplot pertama menampilkan hasil clustering dari Mean Shift.
- Setiap cluster akan ditandai dengan simbol yang berbeda, yaitu "x", "o", dan "^".
- Setiap titik data akan diberi warna sesuai dengan label cluster yang diberikan oleh Mean Shift.
- Pusat dari setiap cluster akan ditandai dengan simbol yang lebih besar dan berwarna hitam.

2. K-Means:
- Subplot kedua menampilkan hasil clustering dari K-Means.
- Seperti pada Mean Shift, setiap cluster akan ditandai dengan simbol yang berbeda.
- Setiap titik data akan diberi warna sesuai dengan label cluster yang diberikan oleh K-Means.
- Pusat dari setiap cluster akan ditandai dengan simbol yang lebih besar dan berwarna hitam.

Kedua subplot akan menunjukkan pola-pola clustering yang dihasilkan oleh masing-masing algoritma. Anda akan dapat melihat bagaimana setiap algoritma membagi data ke dalam cluster-cluster yang berbeda, serta lokasi dari pusat-pusat cluster tersebut.
