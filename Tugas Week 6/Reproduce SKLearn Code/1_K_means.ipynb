Tugas 6 K-means Clustering

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

from sklearn import datasets
from sklearn.cluster import KMeans

np.random.seed(5)

# Load dataset Iris
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Inisialisasi estimator K-Means
estimators = [
    ("k_means_iris_5", KMeans(n_clusters=5)),
    ("k_means_iris_3", KMeans(n_clusters=3)),
    ("k_means_iris_bad_init", KMeans(n_clusters=3, n_init=1, init="random")),
]

# Penjelasan:
# Estimator K-Means diinisialisasi dengan tiga konfigurasi berbeda:
# 1. 5 clusters
# 2. 3 clusters
# 3. 3 clusters dengan inisialisasi yang buruk

# Pembuatan plot
fig = plt.figure(figsize=(15, 5))

for idx, (name, est) in enumerate(estimators, 1):
    title = name
    ax = fig.add_subplot(1, 3, idx, projection='3d')
    est.fit(X)
    labels = est.labels_

    # Plot scatter untuk data
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=labels.astype(float), edgecolor='k')

    # Plot scatter untuk pusat cluster
    ax.scatter(est.cluster_centers_[:, 0], est.cluster_centers_[:, 1], est.cluster_centers_[:, 2], c='red', marker='*', s=300)

    # Penjelasan:
    # Data dipisahkan berdasarkan cluster yang dihasilkan oleh K-Means.
    # Pusat cluster ditandai dengan bintang merah.

    # Penamaan sumbu
    ax.set_xlabel('Sepal length')
    ax.set_ylabel('Sepal width')
    ax.set_zlabel('Petal length')

    # Penamaan judul
    ax.set_title(title)

plt.show()
```
