# Visualize the Data

ChatGPT Prompt: Beri kode untuk menampilkan pie diagram chart untk fitur Age dengan min 20, max 69

```python
import matplotlib.pyplot as plt

# Memfilter baris berdasarkan rentang usia (20-69)
min_age = 20
max_age = 69

# Menghitung jumlah observasi di setiap kategori usia
bins = [min_age, 30, 40, 50, 60, max_age]
labels = ['20-30', '31-40', '41-50', '51-60', '61+']

df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

age_counts = df['Age_group'].value_counts()

# Memplot diagram pie
plt.figure(figsize=(8, 6))
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Usia (20-69 tahun)')
plt.axis('equal')  # Membuat lingkaran menjadi proporsional
plt.show()
```

GPT Prompt: beri kode untuk menampilkan baris data yang memiliki Age '51-60', '60

```python
# Memfilter baris dengan kelompok usia '51-60' dan '61+'
age_51_to_60 = df.loc[df['Age_group'] == '51-60']
age_60_above = df.loc[df['Age_group'] == '61+']

# Menampilkan data untuk kelompok usia '51-60'
print("Data untuk kelompok usia '51-60':")
print(age_51_to_60)

# Menampilkan data untuk kelompok usia '61+'
print("\nData untuk kelompok usia '61+':")
print(age_60_above)
```

GPT Prompt: ubah ke tipe data numerik

```python
# List fitur yang akan diubah menjadi numerik
fitur_numerik = ['Total experience',
                 'Yearly brutto salary',
                 'Yearly bonus and stocks',
                 'Kurzarbeit hours per week',
                 'Vacation days']

# Mengonversi setiap fitur menjadi numerik
for fitur in fitur_numerik:
    df[fitur] = pd.to_numeric(df[fitur], errors='coerce')
```

GPT Prompt: buatkan kode untuk visualisasi fitur numerik

```python
for feature in fitur_numerik:
    plt.figure(figsize=(8, 6))
    df[feature].plot(kind='hist', bins=100, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```
