```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for a HuggingFace course my whole life.")
```

Code di atas menggunakan Transformers untuk membuat sebuah pipeline untuk analisis sentimen dalam bahasa Inggris. Berikut langkah-langkahnya:

1. from transformers import pipeline: Ini mengimpor fungsi pipeline dari pustaka Transformers.

2. classifier = pipeline("sentiment-analysis"): Ini membuat sebuah pipeline untuk analisis sentimen. Dengan menggunakan string "sentiment-analysis" sebagai argumen, pipeline akan menggunakan model yang telah dilatih sebelumnya untuk melakukan analisis sentimen.

3. classifier("I've been waiting for a HuggingFace course my whole life."): Ini mengambil teks sebagai masukan dan menggunakan model yang telah dilatih sebelumnya untuk menganalisis sentimen teks tersebut. Hasilnya adalah label sentimen (misalnya, "POSITIVE" atau "NEGATIVE") dan skor sentimen yang menunjukkan seberapa kuatnya sentimen tersebut dalam teks yang diberikan.

```python
classifier(
    ["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"]
)
```

Code tersebut memanfaatkan model yang sudah dilatih untuk melakukan analisis sentimen pada dua teks yang diberikan secara bersamaan. Dengan menyediakan list teks sebagai input, kita dapat menganalisis sentimen dari masing-masing teks dalam satu kali pemanggilan fungsi. Berikut adalah langkah-langkahnya:

1. "I've been waiting for a HuggingFace course my whole life.": Ini adalah teks pertama yang akan dianalisis sentimennya. Teks ini mungkin akan dianggap positif karena menyatakan antusiasme terhadap kursus HuggingFace.

2. "I hate this so much!": Ini adalah teks kedua yang akan dianalisis sentimennya. Teks ini mungkin akan dianggap negatif karena menyatakan kebencian.

Dengan menggunakan classifier yang sudah didefinisikan sebelumnya, kita dapat langsung memanggilnya dengan list teks tersebut sebagai argumen, dan hasilnya akan berupa list label sentimen dan skor sentimen untuk masing-masing teks dalam list tersebut.

Hasil output tersebut menunjukkan analisis sentimen untuk dua teks yang diberikan:

Teks pertama ("I've been waiting for a HuggingFace course my whole life.") memiliki label sentimen "POSITIVE" dengan skor sentimen sebesar 0.9598. Ini menandakan bahwa teks tersebut memiliki sentimen positif.

Teks kedua ("I hate this so much!") memiliki label sentimen "NEGATIVE" dengan skor sentimen sebesar 0.9995. Ini menandakan bahwa teks tersebut memiliki sentimen negatif dengan tingkat kepercayaan yang sangat tinggi.

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

Code tersebut menggunakan model zero-shot classification dari Transformers untuk melakukan klasifikasi teks tanpa memerlukan label kelas tertentu. Berikut adalah detailnya:

1. "This is a course about the Transformers library": Ini adalah teks yang akan diklasifikasikan.

2. candidate_labels=["education", "politics", "business"]: Ini adalah daftar label kandidat yang mungkin cocok dengan teks yang diberikan. Dalam hal ini, label kandidat adalah "education" (pendidikan), "politics" (politik), dan "business" (bisnis).

Dengan menggunakan classifier yang telah didefinisikan sebelumnya, kita memanggilnya dengan teks yang akan diklasifikasikan dan daftar label kandidat sebagai argumen. Hasilnya adalah label yang dipilih untuk teks tersebut berdasarkan kandidat label yang diberikan.

Output tersebut memberikan hasil klasifikasi zero-shot untuk teks yang diberikan:

- sequence: Ini adalah teks yang diberikan sebagai input.

- labels: Ini adalah label-label yang diberikan sebagai kandidat kelas yang mungkin cocok dengan teks yang diberikan. Urutan label ini sesuai dengan urutan skor yang terkait.

- scores: Ini adalah skor yang menunjukkan seberapa cocok setiap label dengan teks yang diberikan. Semakin tinggi skor, semakin cocok label tersebut dengan teks yang diberikan. Dalam hal ini, label "education" memiliki skor tertinggi, diikuti oleh "business", dan "politics" memiliki skor terendah.

```python
from transformers import pipeline

generator = pipeline("text-generation")
generator("In this course, we will teach you how to")
```

Code tersebut menggunakan model generasi teks dari Transformers untuk menghasilkan teks berlanjut dari awal yang diberikan. Berikut adalah detailnya:

"In this course, we will teach you how to": Ini adalah teks awal yang akan digunakan sebagai dasar untuk menghasilkan teks lanjutan.
Dengan menggunakan generator yang telah didefinisikan sebelumnya, kita memanggilnya dengan teks awal yang diberikan sebagai argumen. Hasilnya adalah teks yang dihasilkan berdasarkan konteks awal yang diberikan.

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
generator(
    "In this course, we will teach you how to",
    max_length=30,
    num_return_sequences=2,
)
```

Code tersebut menggunakan model generasi teks GPT-2 dari Transformers untuk menghasilkan multiple teks berlanjut dari awal yang diberikan. Berikut adalah detailnya:

1. "In this course, we will teach you how to": Ini adalah teks awal yang akan digunakan sebagai dasar untuk menghasilkan teks lanjutan.

2. max_length=30: Parameter ini menentukan panjang maksimum dari teks yang dihasilkan. Dalam hal ini, maksimal 30 token akan dihasilkan.

3. num_return_sequences=2: Parameter ini menentukan jumlah teks yang akan dihasilkan. Dalam hal ini, dua teks akan dihasilkan.

Dengan menggunakan generator yang telah didefinisikan sebelumnya, kita memanggilnya dengan teks awal dan parameter-parameter yang diberikan. Hasilnya adalah dua teks yang dihasilkan berdasarkan konteks awal yang diberikan.

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.", top_k=2)
```

Code tersebut menggunakan model BERT dari Transformers untuk melakukan pengisian kata tersembunyi (fill-mask) dalam sebuah teks. Berikut adalah detailnya:

1. "This course will teach you all about <mask> models.": Ini adalah teks yang mengandung sebuah tempat untuk pengisian kata tersembunyi (mask). Di sini, <mask> adalah tempat untuk mengisi kata yang hilang.

2. top_k=2: Parameter ini menentukan jumlah kandidat kata yang paling mungkin untuk mengisi tempat yang tersembunyi dalam teks. Dalam hal ini, dua kandidat kata yang paling mungkin akan ditampilkan.

Dengan menggunakan unmasker yang telah didefinisikan sebelumnya, kita memanggilnya dengan teks yang mengandung tempat untuk pengisian kata tersembunyi dan parameter-parameter yang diberikan. Hasilnya adalah dua kandidat kata yang paling mungkin untuk mengisi tempat yang tersembunyi dalam teks tersebut.

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")
```

Code tersebut menggunakan model NER (Named Entity Recognition) dari Transformers untuk menemukan entitas berupa nama dalam sebuah teks. Berikut adalah detailnya:

1. "My name is Sylvain and I work at Hugging Face in Brooklyn.": Ini adalah teks di mana model akan mencari entitas berupa nama.

2. grouped_entities=True: Parameter ini mengatur apakah entitas yang ditemukan akan dikelompokkan bersama atau tidak. Dalam hal ini, entitas yang ditemukan akan dikelompokkan berdasarkan jenisnya.

Dengan menggunakan ner yang telah didefinisikan sebelumnya, kita memanggilnya dengan teks yang diberikan sebagai argumen. Hasilnya adalah entitas-entitas berupa nama yang ditemukan dalam teks tersebut, yang dikelompokkan berdasarkan jenisnya.

```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I work?",
    context="My name is Sylvain and I work at Hugging Face in Brooklyn",
)
```

Code tersebut menggunakan model Question Answering dari Transformers untuk menjawab pertanyaan yang diajukan berdasarkan konteks yang diberikan. Berikut adalah detailnya:

1. "Where do I work?": Ini adalah pertanyaan yang diajukan.

2. "My name is Sylvain and I work at Hugging Face in Brooklyn": Ini adalah konteks atau teks di mana model akan mencari jawaban atas pertanyaan yang diajukan.

Dengan menggunakan question_answerer yang telah didefinisikan sebelumnya, kita memanggilnya dengan pertanyaan dan konteks yang diberikan sebagai argumen. Hasilnya adalah jawaban atas pertanyaan tersebut yang ditemukan dalam konteks yang diberikan.

```python
from transformers import pipeline

summarizer = pipeline("summarization")
summarizer(
    """
    America has changed dramatically during recent years. Not only has the number of
    graduates in traditional engineering disciplines such as mechanical, civil,
    electrical, chemical, and aeronautical engineering declined, but in most of
    the premier American universities engineering curricula now concentrate on
    and encourage largely the study of engineering science. As a result, there
    are declining offerings in engineering subjects dealing with infrastructure,
    the environment, and related issues, and greater concentration on high
    technology subjects, largely supporting increasingly complex scientific
    developments. While the latter is important, it should not be at the expense
    of more traditional engineering.

    Rapidly developing economies such as China and India, as well as other
    industrial countries in Europe and Asia, continue to encourage and advance
    the teaching of engineering. Both China and India, respectively, graduate
    six and eight times as many traditional engineers as does the United States.
    Other industrial countries at minimum maintain their output, while America
    suffers an increasingly serious decline in the number of engineering graduates
    and a lack of well-educated engineers.
"""
)
```

Dengan menggunakan summarizer yang telah didefinisikan sebelumnya, kita memanggilnya dengan teks yang akan dirangkum sebagai argumen. Hasilnya adalah rangkuman dari teks tersebut.

```python
from transformers import pipeline

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")
translator("Ce cours est produit par Hugging Face.")
```
