```python
from datasets import load_dataset
from transformers import AutoTokenizer, DataCollatorWithPadding

raw_datasets = load_dataset("glue", "mrpc")
checkpoint = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)


def tokenize_function(example):
    return tokenizer(example["sentence1"], example["sentence2"], truncation=True)


tokenized_datasets = raw_datasets.map(tokenize_function, batched=True)
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

Kode di atas memuat dataset "MRPC" dari GLUE menggunakan datasets library dan menginisialisasi tokenizer dari model "bert-base-uncased". Fungsi tokenize_function digunakan untuk menggabungkan dan men-tokenisasi dua kalimat (sentence1 dan sentence2) dari setiap contoh dalam dataset, dengan memotong teks yang terlalu panjang. Dataset kemudian dipetakan melalui fungsi ini untuk menghasilkan dataset yang telah di-tokenisasi. Terakhir, DataCollatorWithPadding diinisialisasi dengan tokenizer untuk memastikan setiap batch memiliki panjang yang sama selama proses pelatihan, dengan menambahkan padding sesuai kebutuhan.

```python
from transformers import TrainingArguments

training_args = TrainingArguments("test-trainer")
```

Kode tersebut menginisialisasi objek TrainingArguments dari library Transformers, yang digunakan untuk menentukan berbagai parameter dan konfigurasi pelatihan model. Dalam hal ini, objek training_args dibuat dengan satu argumen yaitu "test-trainer", yang menentukan direktori output di mana hasil pelatihan seperti model dan log akan disimpan. Objek ini mencakup berbagai pengaturan seperti jumlah epoch, ukuran batch, tingkat pembelajaran, dan lainnya, yang dapat disesuaikan lebih lanjut untuk mengontrol proses pelatihan model secara rinci.

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(checkpoint, num_labels=2)
```

Kode ini menginisialisasi model untuk tugas klasifikasi teks menggunakan model "bert-base-uncased" dari library Transformers. AutoModelForSequenceClassification diinstansiasi dengan checkpoint bert-base-uncased dan parameter num_labels disetel ke 2, yang berarti model ini akan digunakan untuk tugas klasifikasi dengan dua label (misalnya, untuk klasifikasi biner seperti apakah dua kalimat memiliki arti yang sama atau tidak). Model ini siap digunakan untuk pelatihan atau evaluasi pada dataset yang telah dipersiapkan sebelumnya.

```python
from transformers import Trainer

trainer = Trainer(
    model,
    training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    data_collator=data_collator,
    tokenizer=tokenizer,
)
```

Kode tersebut menginisialisasi objek Trainer dari library Transformers untuk menangani pelatihan dan evaluasi model. Objek Trainer diinstansiasi dengan beberapa parameter:

- model: model klasifikasi teks yang telah diinisialisasi sebelumnya.
- training_args: objek TrainingArguments yang mengatur parameter pelatihan.
- train_dataset: dataset pelatihan yang telah di-tokenisasi (tokenized_datasets["train"]).
- eval_dataset: dataset evaluasi yang telah di-tokenisasi (tokenized_datasets["validation"]).
- data_collator: objek DataCollatorWithPadding yang memastikan setiap batch memiliki panjang yang sama dengan menambahkan padding.
- tokenizer: tokenizer yang digunakan untuk memproses teks.

Dengan pengaturan ini, Trainer akan menangani proses pelatihan, evaluasi, dan pemrosesan batch secara otomatis.

```python
trainer.train()
```

Pemanggilan trainer.train() akan memulai proses pelatihan model menggunakan semua konfigurasi yang telah diatur sebelumnya. Ini mencakup melatih model dengan dataset pelatihan (train_dataset), melakukan evaluasi menggunakan dataset validasi (eval_dataset), dan iterasi melalui jumlah epoch yang ditentukan dalam objek training_args. Selama proses pelatihan, Trainer akan mengatur pembelajaran, perhitungan gradien, dan penyesuaian bobot model berdasarkan performa yang diamati selama evaluasi. Setelah selesai, Anda dapat mengevaluasi hasilnya menggunakan metrik yang relevan atau menyimpan model yang terlatih untuk digunakan di masa depan.

```python
predictions = trainer.predict(tokenized_datasets["validation"])
print(predictions.predictions.shape, predictions.label_ids.shape)
```

Pada kode di atas, trainer.predict(tokenized_datasets["validation"]) digunakan untuk melakukan prediksi menggunakan model yang telah dilatih terhadap dataset validasi (validation) yang telah di-tokenisasi (tokenized_datasets["validation"]).

Hasil prediksi disimpan dalam objek predictions, yang kemudian mencetak bentuk dari predictions.predictions dan predictions.label_ids.

1. predictions.predictions.shape memberikan bentuk dari hasil prediksi yang diperoleh dari dataset validasi. Jumlah baris dalam predictions.predictions sesuai dengan jumlah sampel dalam dataset validasi, dan jumlah kolomnya sesuai dengan jumlah label yang diprediksi.

2. predictions.label_ids.shape memberikan bentuk dari label sebenarnya (ground truth) yang ada dalam dataset validasi. Jumlah elemen dalam predictions.label_ids sama dengan jumlah sampel dalam dataset validasi.

```python
import numpy as np

preds = np.argmax(predictions.predictions, axis=-1)
```

Kode di atas menggunakan NumPy untuk melakukan operasi argmax pada hasil prediksi yang diperoleh dari trainer.predict(tokenized_datasets["validation"]). Mari kita bahas langkah-langkahnya:

- predictions.predictions: Ini adalah array yang berisi hasil prediksi dari model untuk dataset validasi. Setiap baris mewakili satu sampel dalam dataset validasi, dan setiap kolom mungkin mewakili skor prediksi untuk setiap kelas (dalam kasus ini, untuk tugas klasifikasi biner, biasanya dua kelas).

- np.argmax(predictions.predictions, axis=-1): Operasi argmax digunakan untuk menentukan indeks dari nilai terbesar dalam setiap baris dari predictions.predictions. Dalam konteks klasifikasi, ini berarti kita memilih kelas dengan nilai prediksi tertinggi sebagai prediksi akhir untuk setiap sampel. Parameter axis=-1 digunakan untuk melakukan argmax di sepanjang dimensi terakhir dari array, yang dalam kasus ini adalah dimensi kolom dari predictions.predictions.

- preds: Variabel ini menyimpan hasil dari operasi argmax, yang berisi indeks kelas yang diprediksi untuk setiap sampel dalam dataset validasi. Misalnya, jika hasil argmax adalah 0, itu berarti model memprediksi sampel tersebut sebagai kelas pertama, dan jika hasilnya adalah 1, itu berarti model memprediksi sampel tersebut sebagai kelas kedua (dalam kasus biner).

```python
import evaluate

metric = evaluate.load("glue", "mrpc")
metric.compute(predictions=preds, references=predictions.label_ids)
```

```python
def compute_metrics(eval_preds):
    metric = evaluate.load("glue", "mrpc")
    logits, labels = eval_preds
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

```python
training_args = TrainingArguments("test-trainer", evaluation_strategy="epoch")
model = AutoModelForSequenceClassification.from_pretrained(checkpoint, num_labels=2)

trainer = Trainer(
    model,
    training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    data_collator=data_collator,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```
